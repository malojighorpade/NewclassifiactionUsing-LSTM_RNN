```python
X=df.drop('label',axis=1)
```

```python
y=df['label']
```

```python
X.shape
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
voc_size=7000
```

```python
messages=X.copy()
```

```python
messages['text'][1]
```

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
nltk.download('stopwords')
```

```python
corpus = []
for i in range(0, len(messages)):

    review = re.sub('[^a-zA-Z0-9]', ' ', messages['text'][i])
    review=review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
onehot_repr=[one_hot(i,voc_size)for i in corpus]
onehot_repr
```

```python
corpus[9]
```

```python
onehot_repr[9]
```

```python
max_lenght=30
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=max_lenght)
print(embedded_docs)

```

```python
embedded_docs[90]
```

```python
model = Sequential()
embedding_vector_features=40
model.add(Embedding(voc_size, embedding_vector_features,input_shape=(max_lenght,)))
model.add(LSTM(100))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
import numpy as np
X_final1=np.array(embedded_docs)
y_final2=np.array(y)
```

```python
X_final1.shape,y_final2.shape
```

```python
from sklearn import model_selection
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_final1, y_final2, test_size=0.33, random_state=50
)
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

```python
model.evaluate(X_test, y_test)
```
